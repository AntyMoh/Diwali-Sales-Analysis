# Exploratory Data Analysis

Gender

```python
sns.countplot(x='Gender', data=sales)
```

```python
ax=sns.countplot(x='Gender', data=sales)

for bars in ax.containers:
    ax.bar_label(bars)
```

From above graph we deduce that most of the buyers are females and even the purchasing power of females are greater than man

# Age

```python
sns.countplot(x='Age Group', data=sales,hue='Gender')
```

```python
ax=sns.countplot(x='Age Group', data=sales,hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Total Amount vs Age Group
sales_age=sales.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending= False)
sns.barplot(x='Age Group',y='Amount',data=sales_age)
```

From the Above graphs we saw that most of the buyers are of age 26-35 who are Female

# State

```python
#Total No. of orders from top 10 states
sales_state=sales.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
ax=sns.barplot(x='State', y='Orders',data=sales_state)


#Dont want to set column ID as Index
```

```python
#Total Amount/sales from top 10 states
sales_state=sales.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)}) #to incerse space between the x labels of bars
ax=sns.barplot(x='State', y='Amount',data=sales_state)
for bars in ax.containers:
    ax.bar_label(bars)
```

From the Above Graph it is derixed that most of the orders are from the states of Utar Pradesh, Maharashtra, and Karnataka.Similary the Total sales is Maximum in the states of UP, Maharashtra and Karnataka 

# Marital Status

```python
ax=sns.countplot(x='Marital_Status', data=sales)
sns.set(rc={'figure.figsize':(6,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_ms= sales.groupby(['Marital_Status','Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(x='Marital_Status', y='Amount',data=sales_ms,hue='Gender' )
```

From the graphs we see most of buyers are married women and they have high purchasing power

# Occuaption

```python
sns.set(rc={'figure.figsize':(20,5)})
ax=sns.countplot(x='Occupation', data=sales)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Total sales by various Occuaptions
sales_occu= sales.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='Occupation', y='Amount',data=sales_occu )
```

From the above graph we see that most of the buyers are working in IT, Healthcare and Aviation

# Product Category

```python
sns.set(rc={'figure.figsize':(20,5)})
ax=sns.countplot(x='Product_Category', data=sales)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Total sales by various Product Category
sales_occu= sales.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(5)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='Product_Category', y='Amount',data=sales_occu )
```

From the graph most of the sales are from Food , Clothing&Apparel and Gadgets

```python
#Top selling products
sales_prod= sales.groupby(['Product_ID'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='Product_ID', y='Amount',data=sales_prod)
```

# conclusion

### Married Women age group 26-35 yrs from UP, Maharastra and Karnataka working in IT, Healthcare and aviation sectors are more likely to buy products from Food , Clothing and Electronics Category
